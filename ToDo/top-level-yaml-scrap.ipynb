```python
config_yaml_path = 'top-level-yaml-test.yaml'
```

```python
import yaml

# Load and resolve references
def resolve_paths(config):
    resolved = {}
    for key, value in config.items():
        if isinstance(value, str):
            try:
                resolved[key] = value.format(**resolved)
            except KeyError:
                resolved[key] = value  # Leave unresolved for now
        else:
            resolved[key] = value
    # Try a second pass to resolve remaining keys
    for key, value in resolved.items():
        if isinstance(value, str):
            resolved[key] = value.format(**resolved)
    return resolved

# Load YAML
with open(config_yaml_path, "r") as f:
    raw_config = yaml.safe_load(f)

config = resolve_paths(raw_config)

# Access resolved paths
#print(config["output_dir"])  # -> /home/user/project/data/output
#print(f'''{config}''')  # -> /home/user/project/data/output


for key, value in config.items():
    print(f"{key}: {value}")
```
